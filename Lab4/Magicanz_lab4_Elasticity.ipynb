# **Introduction**

The Elasticity equations take the form

$-\nabla \cdot \sigma = f$

together with suitable boundary conditions.

To derive the weak form of the equations, multiply the equation by $v\in V$, and then integrate over the domain $\Omega$ and use Green's formula
$
(-\nabla \cdot \sigma ,v) = (\nabla \sigma , \nabla v)
- <\sigma \cdot n, v>_{\Gamma}
$

To model

We seek a finite element approximation $u\in V$ such that

$(\nabla \sigma,\nabla v) - <\sigma \cdot n, v>_{\partial \Omega} = (f,v)$

for all test functions $v \in V$.



# **Method**

**Define domain and mesh**

```python
# Define rectangular domain
L = 4.0
H = 2.0

# Define subdomains (for boundary conditions)
class Left(SubDomain):
    def inside(self, x, on_boundary):
        return near(x[0], 0.0)

class Right(SubDomain):
    def inside(self, x, on_boundary):
        return near(x[0], L)

class Lower(SubDomain):
    def inside(self, x, on_boundary):
        return near(x[1], 0.0)

class Upper(SubDomain):
    def inside(self, x, on_boundary):
        return near(x[1], H)

left = Left()
right = Right()
lower = Lower()
upper = Upper()

# Generate mesh (examples with and without a hole in the mesh)
resolution = 32
mesh = generate_mesh(Rectangle(Point(0.0,0.0), Point(L,H)) - Circle(Point(1.5,0.25*H),0.2) - Circle(Point(0.5,0.5*H),0.2) - Circle(Point(2.0,0.75*H),0.2), resolution)

# Local mesh refinement (specified by a cell marker)
no_levels = 0
for i in range(0,no_levels):
  cell_marker = MeshFunction("bool", mesh, mesh.topology().dim())
  for cell in cells(mesh):
    cell_marker[cell] = False
    p = cell.midpoint()
    if p.distance(Point(1.5, 0.5)) < 1.0:
        cell_marker[cell] = True
  mesh = refine(mesh, cell_marker)

# Define mesh functions (for boundary conditions)
boundaries = MeshFunction("size_t", mesh, mesh.topology().dim()-1)
boundaries.set_all(0)
left.mark(boundaries, 1)
right.mark(boundaries, 2)
lower.mark(boundaries, 3)
upper.mark(boundaries, 4)

plt.figure()
plot(mesh)
plt.show()
```

**Define finite element approximation spaces**

```python
# Generate finite element space
VE = VectorElement("CG", mesh.ufl_cell(), 1)

V = FunctionSpace(mesh, VE)

# Define trial and test functions
u = TrialFunction(V)
v = TestFunction(V)

d = Function(V)
```

**Define boundary conditions**

```python
# Define boundary conditions
class DirichletBoundaryLower(SubDomain):
    def inside(self, x, on_boundary):
        return on_boundary and near(x[1], 0.0)

class DirichletBoundaryUpper(SubDomain):
    def inside(self, x, on_boundary):
        return on_boundary and near(x[1], H)

class DirichletBoundaryLeft(SubDomain):
    def inside(self, x, on_boundary):
        return on_boundary and near(x[0], 0.0)

class DirichletBoundaryRight(SubDomain):
    def inside(self, x, on_boundary):
        return on_boundary and near(x[0], L)

class DirichletBoundaryObjects(SubDomain):
    def inside(self, x, on_boundary):
        return on_boundary and (not near(x[0], 0.0)) and (not near(x[0], L)) and (not near(x[1], 0.0)) and (not near(x[1], H))

dbc_lower = DirichletBoundaryLower()
dbc_upper = DirichletBoundaryUpper()
dbc_left = DirichletBoundaryLeft()
dbc_right = DirichletBoundaryRight()
dbc_objects = DirichletBoundaryObjects()

bcu_left0 = DirichletBC(V.sub(0), 0.0, dbc_left)
bcu_left1 = DirichletBC(V.sub(1), 0.0, dbc_left)
bcu_right0 = DirichletBC(V.sub(0), 0.0, dbc_right)
bcu_right1 = DirichletBC(V.sub(1), 0.0, dbc_right)
bcu_upper0 = DirichletBC(V.sub(0), 0.0, dbc_upper)
bcu_upper1 = DirichletBC(V.sub(1), 0.0, dbc_upper)
bcu_lower0 = DirichletBC(V.sub(0), 0.0, dbc_lower)
bcu_lower1 = DirichletBC(V.sub(1), 0.0, dbc_lower)
bcu_obj0 = DirichletBC(V.sub(0), 1.5, dbc_objects)
bcu_obj1 = DirichletBC(V.sub(1), 0.0, dbc_objects)

#bcu = [bcu_in0, bcu_in1, bcu_upp0, bcu_upp1, bcu_low0, bcu_low1, bcu_obj0, bcu_obj1]
bcu = [bcu_left0, bcu_left1, bcu_right0, bcu_right1, bcu_upper0, bcu_upper1, bcu_lower0, bcu_lower1, bcu_obj0, bcu_obj1]
```

# **Results**

**Define and solve variational problem**

```python
f = Expression(("0.0","0.0"), element = V.ufl_element())

dim = u.geometric_dimension()
E = 3*1e10
nu = 0.1
mu = E*0.5/(1+nu)
lambda_ = nu*E/((1.0+nu)*(1.0-2.0*nu))

def epsilon(u):
    return 0.5*(grad(u) + grad(u).T)

def sigma(u):
    #return 2.0*mu*epsilon(u)
    return lambda_*div(u)*Identity(dim) + 2.0*mu*epsilon(u)

# Define variational problem on residual form: r(u,v) = 0
residual = ( inner(sigma(u), epsilon(v))*dx - inner(f, v)*dx )

au = lhs(residual)
Lu = rhs(residual)

A = assemble(au)
b = assemble(Lu)

[bc.apply(A, b) for bc in bcu]
[bc.apply(d.vector()) for bc in bcu]

solve(A, d.vector(), b, "bicgstab", "default")
```

**Visualize solution and export files**

```python
u1 = project(d, V)

# Plot solution
plt.figure()
plot(u1, title="Displacement")

plt.show()
```

**Update mesh**

# **Mesh Deformation**
Depending on factors like Young's modulus $E$, Poisson's ratio $ν$, and the movement of the cylinders the mesh deforms in different ways.

![normal](https://i.imgur.com/XXdYITJ.png)

Originally, the movement is $0.5$ units to the right, with $E=10^{10}$ and $\nu = 0.3$, which can be seen in the figure above. This is the base mesh deformation that we will consider. It can be seen that it resists the deformation quite well and does not unevenly bend, with most deformations being done in the horizontal direction. From looking at some tables, these values could pertain to some kind of wood.

## **Rubber**
![rubber](https://i.imgur.com/I07hkzr.png)

By setting $E=10^8, \nu=0.48$ we should get something that acts like rubber. As is seen in the image above, that seems realistic. The deformation flows together between the cylinders in a smooth manner, and it can be seen that there have been a lot of changes around the cylinders, wheras in the original scenario most of the mesh seems to have retained its original shape.

## **Reinforced Concrete**
![concrete](https://i.imgur.com/lPup6tt.png)

By setting $E=3*10^{10}, \nu=0.1$ we should get something that acts like concrete reinforced and compressed by rebar. As one would expect with concrete, it is quite firm and unbending. This is seen in how even less of the mesh is modified compared to the base scenario. It resists the movement quite well.

## **Diamond**
![diamond](https://i.imgur.com/otTJ4Je.png)

By setting $E=1.22*10^{12}, \nu=0.2$ we should get something that acts like Diamond. Known as one of the hardest materials on earth, we would expect a low deformation, and this is what we see in the images above. They look very similar to the ones for Reinforced Concrete, meaning there is a ceiling to how little the mesh may be deformed with a set movement.

## **Downwards Movement**
![downmove](https://i.imgur.com/LzwgW2c.png)

Instead of changing the material (which will be kept at $E=10^{10}, \nu=0.3$), we now change the movement. In this case, we move it 0.25 units downwards. Since there isn't much room to move, the first obvious thing is that there is less movement in general We can however see that the area underneath the bottom cylinder gets very squished unto the boundary, something that may be problematic. We can also see here that the triangles at the top are the ones that get most deformed compared to earlier when the left triangles got the most deformation applied.

## **Far Rightward Movement**
![rightmove](https://i.imgur.com/W8dKoNI.png)

If we move the cylinders much further to the right than usual, we get some interesting results. In the image above, the cylinders have been moved 1.5 units to the right. This has caused the triangles to the right to wrap in over themselves which is something that is very problematic in a 2D setting, but that may be possible to work with if we had a third dimension. We can also see that the stretching of the triangles to the left has caused some of them too to wrap in on themselvs in a few lines, one connected to each cylinder.
